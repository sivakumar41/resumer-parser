# Importing the needed libraries

```python
import spacy
import json
from spacy.tokens import DocBin
from sklearn.model_selection import train_test_split
```

```python
#reading the json data
with open('resumeparser.tds.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
print("The lenght of the total json data is",len(data))
```

```python
#spliting the total data into training and testing data
training_data, testing_data = train_test_split(data, test_size=0.3, random_state=42)
print("The lenght of the training data is ",len(training_data))
print("The lenght of the testing data is ",len(testing_data))
```

```python
#creating the nlp blank 
```

```python
ty=[]
for item in training_data:
    ty.append(type(item))
print(set(ty))
#here we observe that we have only the dict type of collection in the dataset
```

```python
# preprocessing the training data
nlp=spacy.blank("en")
db=DocBin()  # object to the DocBin()
skipped=0
total=0
org_lables=["NAME","EMAIL","ADDRESS","GRADUATION","PERCENTAGE","PROJECTS","JOB ROLE","SKILLS","WORK EXPERIENCE"," TECHNOLOGIES","CERTIFICATIONS","ACHIEVEMENTS","EXTRA CURRICULAR","ONLINE PROFILES"]
for item in training_data:
    if isinstance(item, dict):
        text = item.get('text', '')
        entities = item.get('entities', [])
    elif isinstance(item, tuple):
        text, entities = item
    else:
        continue
    doc = nlp.make_doc(text)
    valid_ents = []

    try:
        for start, end, label in entities:
            if label in org_lables:
                
                span = doc.char_span(start, end, label=label, alignment_mode="contract")
                total += 1
                if (span is None or span.text.startswith(" ") or span.text.endswith(" ") or span.text != span.text.strip() ):
                    ent_text = text[start:end]
                    print(f"⚠  Skipping Entity : {text[0:30]}... {ent_text}")
                    skipped += 1
                else:
                    valid_ents.append(span)

        doc.ents = valid_ents
        db.add(doc)
    except Exception as ex:
        print("⚠ ", ex)
        skipped += 1

db.to_disk("training_data.spacy")
print("Total Number of skiped entities are-->>",skipped)

    
        
```

```python
#The above same process for the testing data too
nlp = spacy.blank("en")
db = DocBin()


for item in testing_data:
    if isinstance(item, dict):
        text = item.get('text', '')
        entities = item.get('entities', [])
    elif isinstance(item, tuple):
        text, entities = item
    else:
        continue

    doc = nlp.make_doc(text)
    valid_ents = []

    try:
        for start, end, label in entities:
            if label in org_lables:
                span = doc.char_span(start, end, label=label, alignment_mode="contract")
                total += 1
                if (
                    span is None
                    or span.text.startswith(" ")
                    or span.text.endswith(" ")
                    or span.text != span.text.strip()
                ):
                    ent_text = text[start:end]
                    print(f"⚠  Skipping Entity : {text[0:30]}... {ent_text}")
                    skipped += 1
                else:
                    valid_ents.append(span)

        doc.ents = valid_ents
        db.add(doc)
    except Exception as ex:
        print("⚠ ", ex)
        skipped += 1

db.to_disk("testing_data.spacy")
```

```python
training_data_len=len(training_data)
print("✨ Results:")
print(f"Total Training Data : {training_data_len}")

print(f"Failed to Convert to spacy : {skipped} / {total} = {skipped/total*100} %")
```
